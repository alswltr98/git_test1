# 넘파이(umpy)

```python
# 필요한 라이브러리 임포트
import numpy as np
```

## 넘파이 배열과 파이썬 리스트 비교

```python
# 파이썬 리스트 자료 구조 데이터 생성
list_data = [1,2,3]
print(f'파이썬 리스트: {list_data}')

print('-'*80)

# 넘파이 배열 생성
array_data = np.array(list_data)
print(f'넘파이 배열로 변한 후 데이터: {array_data}')
```

```python
# 파이썬 리스트와 넘파이 배열 차이점 확인

## 덧셈 연산

### 파이썬 리스트
result1 = list_data + list_data
print(f'파이썬 리스트에 대한 덧셈 연산의 결과: \n{result1}')

print('-'*80)

### 넘파이 배열
result2 = array_data + array_data
print(f'넘파이 배열에 대한 덧셈 연산의 결과: \n{result2}')

print('-'*80)

## 곱셈 연산

### 파이썬 리스트
result3 = []

for num in list_data:
    result3.append(num*3)

print(f'파이썬 리스트에 대한 곱셈 연산의 결과: \n{result3}')

print('-'*80)

### 넘파이 배열
result4 = array_data * 3
print(f'넘파이 배열에 대한 곱셈 연산의 결과: \n{result4}')

print('-'*80)

## 통계 처리(평균 값 구하기)

### 파이썬 리스트
list_average = sum(list_data) / len(list_data)
print(f'파이썬 리스트에 대한 평균 값: {list_average}')

print('-'*80)

### 넘파이 배열
array_average = array_data.mean()
print(f'넘파이 배열에 대한 평균 값: {array_average}')
```

```python
# 대괄호가 중첩된 리스트 자료 구조 데이터 생성(조건: 성분 원소의 개수 동일)
list_data = [[1,2,3],[4,5,6],[7,8,9]]
print(f'대괄호가 중첩된 리스트 자료 구조 데이터 확인: \n{list_data}')

print('-'*80)

# 넘파이 배열로 변환 -> 엑셀과 같은 표(table: 행과 열) 구조 생성
array_data = np.array(list_data)
print(f'대괄호가 중첩된 리스트를 넘파이 배열로 변환한 결과: \n{array_data}')
```

## 2차원 배열 생성

```python
# 대괄호가 중첩된 리스트 자료 구조 데이터 생성(조건: 성분 원소의 개수 동일)
list_data = [[1,2,3],[4,5,6]]
print(f'대괄호가 중첩된 리스트 자료 구조 데이터 확인: \n{list_data}')

print('-'*80)

# 넘파이 배열로 변환 -> 2차원 배열 -> 엑셀과 같은 표(table: 행과 열) 구조 생성
array_data = np.array(list_data)
print(f'대괄호가 중첩된 리스트를 넘파이 배열로 변환한 결과: \n{array_data}')
```

## 인덱싱

```python
# 목표: 첫 번째 행, 첫 번째 열 -> 1 추출

## 파이썬 방식

### 1단계: 변수[인덱스] -> 첫 번째 행 추출
result1 = array_data[0]
print(result1)

print('-'*80)

### 2단계: 변수[인덱스] -> 첫 번째 성분(첫 번째 열) 원소
# result2 = result1[0]
result2 = array_data[0][0]
print(result2)

print('-'*80)

## 넘파이 방식: 넘파이의 배열 모양(2행,3열) 이용 -> 변수[행 인덱스, 열 인덱스]
result3 = array_data[0,0]
print(f'숫자 1을 인덱싱한 결과: {result3}')
```

## 새로운 컬럼 추가(파생 변수 생성)

```python
# CTR (Click-Through Rate): 클릭률 컬럼 추가
df['CTR'] = df['Clicks'] / df['Impressions'] * 100
# display(df)

# print('-'*80)

# CPC (Cost Per Click): 클릭당 비용 컬럼 추가
df['CPC'] = df['Cost'] / df['Clicks']
# display(df)

# CVR (Conversion Rate): 전환율 컬럼 추가
df['CVR'] = df['Conversions'] / df['Clicks'] * 100

# CPA (Cost Per Acquisition): 전환당 비용 컬럼 추가
df['CPA'] = df['Cost'] / df['Conversions']

# ROAS (Return On Ad Spend): 광고비 대비 수익률 컬럼 추가
df['ROAS'] = df['Revenue'] / df['Cost'] * 100

# 최종 결과
display(df)
```

## 전용 함수

### query() 함수

#### 단일 조건(문자열)

```python
# Channel 컬럼의 값이 'Facebook'과 같은 경우
df1 = df.query("Channel == 'Facebook'")
display(df1)
```

#### 단일 조건(숫자)

```python
# Revenue 컬럼의 값이 1500000 이상인 경우
df2 = df.query("Revenue >= 1500000")
display(df2)
```

#### 다중 조건(논리 연산자 and, or)

```python
# and 연산자: 모든 조건을 동시에 만족해야 되는 경우
df3 = df.query("Channel == 'Google' and Revenue >= 1500000")
print("Channel이 Google이면서 총 매출이 1500000 이상인 행 추출")
display(df3)
```

```python
# or 연산자: 제시된 조건 중 하나만 만족해도 되는 경우
df4 = df.query("Age_Group == '20s' or Age_Group == '30s'")
print("연령대가 20대이거나 30대인 행 추출")
display(df4)
```

# merge() 함수

```python
# 데이터 불러오기

## 파일 경로 설정
file_path1 = '/content/drive/MyDrive/KDT/employee.csv'
file_path2 = '/content/drive/MyDrive/KDT/salary.csv'

## DataFrame 생성
df_employee = pd.read_csv(file_path1)
df_salary = pd.read_csv(file_path2)

## 결과 확인
display(df_employee)
print('-'*80)
display(df_salary)
```

```python
# how = 'left'

## 왼쪽 데이터프레임 기준으로 결합
df_merge_left = pd.merge(df_employee, df_salary, on='id', how='left')
display(df_merge_left)
```

```python
# how = 'right'

## 오른쪽 데이터프레임 기준으로 결합
df_merge_right = pd.merge(df_employee, df_salary, on='id', how='right')
display(df_merge_right)
```

```python
# how = 'inner'

## 양쪽에 공통으로 존재하는 행만 결합
df_merge_inner = pd.merge(df_employee, df_salary, on='id', how='inner')
display(df_merge_inner)
```

```python
# how = 'outer'

## 한쪽에만 존재하는 행도 포함해서 결합
df_merge_outer = pd.merge(df_employee, df_salary, on='id', how='outer')
display(df_merge_outer)
```
